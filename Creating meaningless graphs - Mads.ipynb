```python
import random as rd
import matplotlib.pyplot as plt

def RandomList(length, start, end):
    rlist = []
    for i in range(length):
        rlist.append(rd.randint(start, end))
    return rlist

length = 10
start = 1
end = 5

plt.plot(RandomList(length, start, end))
plt.title("PLot of nothing")
plt.grid()
```

The function `RandomList()` takes three arguments, defined by the user; `length` being the number of data points the user wishes to generate, `start` and `end` being the interval from which the user wants the random values generated.
This function returns a list of $y$ values which can be plotted against anything. In the above example, the $x$ values are the indecies of the individual points, starting at `0` and ending at `9`.

One could also generate a seried of $x$ values with this function, however one must be careful and select the same integer for `length` when using this function, as the plotting-function will need to pair up each $x$ and $y$ value.
This is shown below.

```python
length = 7
x = RandomList(length, 0,10)
y = RandomList(length, 5,50)

plt.plot(x,y)
plt.xlabel(f"{length} random values from 0 to 10")
plt.ylabel(f"{length} random values from 5 to 50")
plt.title("Another plot of nothing")
plt.grid()
```

As it can be seen above, the line has the possibility of overlapping itself, as the $x$ values are not in ascending order.
This can be dealt with (if needed) by adding `sort()` to the code as shown below.

```python
length = 7
x = RandomList(length, 0,10)
y = RandomList(length, 5,50)
x.sort() # Sorts the x values in ascending order
print(f"the x values are now sorted {x}")

plt.plot(x,y)
plt.xlabel(f"{length} sorted random values from 0 to 10")
plt.ylabel(f"{length} random values from 5 to 50")
plt.title("Yet another plot of nothing")
plt.grid()
```

These examples are very simplistic in nature, for demonstration purposes, and can be taken to much greater lengths.
